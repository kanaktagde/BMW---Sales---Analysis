```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv("/content/bmw.csv")
data.head()
```

```
data.info()
data.describe()
```

```
data.columns
```

```
model_price = data.groupby("model")["price"].mean()
model_price
```

```
plt.figure(figsize=(10,5))
sns.barplot(x=model_price.index, y=model_price.values)

plt.title("Average BMW Price by Model")
plt.xlabel("BMW Model")
plt.ylabel("Average Price")

plt.xticks(rotation=45)
plt.show()
```

```
year_count = data["year"].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=year_count.index, y=year_count.values)

plt.title("BMW Cars by Year")
plt.xlabel("Year")
plt.ylabel("Number of Cars")

plt.show()
```
